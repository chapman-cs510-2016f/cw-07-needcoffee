```python
import juliaPlane as jp
import numpy as np
```

The juliaPlane module, which is a cumulation of several weeks of coding complex planes and related function, is finally complete. The result is a plot of the Julia set defined by the parameter c. The Julia set is closely related to the Mandelbrot set, one of the more well-known fractals, even to the general public.

```python
exPlane1 = jp.JuliaPlane(-2,2,1000,-2,2,1000)
exPlane1.set_f(0)
exPlane1.show()
```

```python
exPlane2 = jp.JuliaPlane(-2,2,1000,-2,2,1000)
exPlane2.set_f(-1)
exPlane2.show()
```

```python
exPlane3 = jp.JuliaPlane(-2,2,1000,-2,2,1000)
exPlane3.set_f(0.3)
exPlane3.show()
```

```python
exPlane4 = jp.JuliaPlane(-2,2,1000,-2,2,1000)
exPlane4.set_f(-1.037 + 0.17j)
exPlane4.show()
```

```python
exPlane5 = jp.JuliaPlane(-2,2,1000,-2,2,1000)
exPlane5.set_f(-0.624 + 0.435j)
exPlane5.show()
```

```python
exPlane6 = jp.JuliaPlane(-2,2,1000,-2,2,1000)
exPlane6.set_f(-0.8 + 0.2j)
exPlane6.show()
```

```python
exPlane7 = jp.JuliaPlane(-2,2,1000,-2,2,1000)
exPlane7.set_f(-0.835 - 0.2321j)
exPlane7.show()
```

The c values for the last two Julia sets were taken from the Julia Set Wikipedia page

```python
phi = (1 + np.sqrt(5)) / 2

exPlane8 = jp.JuliaPlane(-2,2,1000,-2,2,1000)
exPlane8.set_f((phi-2) + (phi-1)*1j)
exPlane8.show()
```

```python
exPlane9 = jp.JuliaPlane(-2,2,1000,-2,2,1000)
exPlane9.set_f(-0.7269+0.1889j)
exPlane9.show()
```
